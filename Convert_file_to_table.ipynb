```python
import pandas as pd
```

```python
with open ("V2_Statistik_Auswertung_202301.txt") as f: 
    stats = f.readlines()
```

```python
stats
```

```python
def convert_to_dict(line):

    first = line.split("]")
    date = first[0].replace("[", "")
    rest = first[1]
    second = rest.split()
    request_type = second[0].replace('"', '')
    request = second[1]
    http = second[2]
    reply1 = second[3]
    reply2 = second[4]
    no_idea = second[5]
    client = second[6]
    java = second[7]
    
    
        
    result = {"date": date, "request_type": request_type, "request": request, "http": http, 'reply1':reply1, 
             'reply2': reply2, 'no_idea': no_idea, 'client': client, 'java':java}
       
    return result
```

```python
result = [convert_to_dict(line) for line in stats]
```

```python
result
```

```python
df = pd.DataFrame(result)
df
```

```python
df[['namensraum', 'id']] = df['request'].str.split('-', 1, expand=True)
```
